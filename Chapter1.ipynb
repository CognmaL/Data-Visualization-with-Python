### ***Bài học rút ra:***
Việc xử lý các phần tử gần như tuân theo quy tắc truy xuất của cấu trúc mảng trong lập trình.

Dưới đây là một số kỹ thuật truy xuất phần tử trên mảng:

**Indexing**

Đối với NumPy, chúng ta có thể xử lý số liệu bằng cách đưa nó về mảng n chiều (ndarrays). Cao cấp hơn, chúng ta dùng cấu trúc dữ liệu `list` trong Python.

```
dataset[0] # index ở vị trí ngoài cùng (đầu tiên) của dữ liệu 1 chiều 
dataset[-1] # index theo thứ tự đảo ngược ở chiều ngoài cùng.
dataset[1, 1] # index phần tử đơn lẻ trong dữ liệu 2 chiều 
dataset[-1, -1] # index thứ tự đảo ngược ở dữ liệu 2 chiều
```

**Slicing**

Kỹ thuật `cắt lát` cũng được hiệu chỉnh trong `list`, kỹ thuật này rất hữu dụng khi thao tác với những bộ dữ liệu lớn
```
dataset[1:3] # dòng 1 và 2
dataset[:2, :2] # tập hợp con 2x2 của dữ liệu
dataset[-1, ::-1] # đảo ngược các phần tử ở dòng cuối cùng
dataset[-5:-1, :6:2] # ở 4 hàng cuối cùng, mọi phần tử khác lên đến index 6
```
**Splitting**

Kỹ thuật `chia nhỏ dữ liệu` hữu dụng trong các đồ thị chỉ yêu cầu vẽ một nửa dữ liệu kiểu chuỗi thời gian để tách dữ liệu ra kiểm tra và train cho các thuật toán ML.

Có hai cách để chia nhỏ dữ liệu, theo chiều ngang và chiều dọc. 
- Chia ngang có thể được thực hiện bằng phương pháp `hsplit`.
- Tách dọc có thể được thực hiện bằng phương pháp `vsplit`:
```
np.hsplit(dataset, (3)) # Chia theo chiều ngang thành 3 danh sách bằng nhau
np.vsplit(dataset, (2)) # Chia theo chiều dọc thành 2 danh sách bằng nhau
```
**Iterating**

`Phép lặp` là cấu trúc dữ liệu trong NumPy. Nó lần lượt đi qua toàn bộ danh sách dữ liệu, truy cập mọi phần tử đơn trong ndarray một lần. Trong những trường hợp dữ liệu có nhiều chiều, sự lặp trở nên rất phức tạp.

`nditer` là một đối tượng iterator đa chiều lặp lại trên một số mảng nhất định:
```
# Lặp lại trên toàn bộ tập dữ liệu (mỗi giá trị trong mỗi hàng)
for x in np.nditer(dataset):
print(x)
```

`ndenumerate` sẽ cung cấp cho chúng ta chính xác chỉ số này, do đó trả về (0, 1) cho giá trị thứ hai trong hàng đầu tiên:

```
#Lặp lại trên toàn bộ tập dữ liệu với các chỉ số khớp với vị trí trong tập dữ liệu
for index, value in np.ndenumerate(dataset):
print(index, value)
```

Do đó, chúng ta có căn cứ để khẳng định với khách hàng rằng tập dữ liệu mà họ đang nắm có phân phối "đẹp"

#### Splitting

Nếu cần một tập dữ liệu nhỏ hơn, chúng ta cũng có thể sử dụng tính năng chia tách để phân chia tập dữ liệu của mình một cách hiệu quả.

Giả sử tập dữ liệu của chúng ta chia như sau: 

```
1, 2, 3, 4, 5, 6          1, 2     3, 4    5, 6          1, 2  
3, 2, 1, 5, 4, 6    =>    3, 2     1, 5    4, 6    =>    3, 2    =>    1, 2
5, 3, 1, 2, 4, 3          5, 3     1, 2    4, 3                        3, 2
1, 2, 2, 4, 1, 5          1, 2     2, 4    1, 5          5, 3
                                                         1, 2
```

Có nghĩa là nó được chia thành 3 phần bằng nhau và sau đó lại lấy phần đầu của lần chia đầu tiên chia thành 2 nửa nhỏ.

Việc phân chia dữ liệu theo chiều ngang có thể được thực hiện bằng phương pháp hsplit. 

Tuy nhiên, ta không thể phân chia tập dữ liệu với số lát = 3 như ví dụ, nó sẽ báo lỗi như sau:

````python
# splitting up our dataset horizontally on indices one third and two thirds
hor_splits = np.hsplit(dataset,(3))
````

Lỗi này cho biết số lượng phần cần chia thành các mảng con đang không bằng nhau.

Để sửa lỗi này, bạn cần chỉnh sửa số lượng phần hoặc chỉ mục bạn đang sử dụng để chia mảng sao cho chia đều.

````python
hor_splits = np.hsplit(dataset,(8))
print("Dataset", dataset.shape)
print("Subset", hor_splits[0].shape)
````

Bây giờ chúng ta cần chia phần ba phần đầu tiên thành hai phần bằng nhau theo chiều dọc. 

Dùng `vsplit`:

````python
# splitting up our dataset vertically on index 2
ver_splits = np.vsplit(hor_splits[0],(2))
````

````python
# requested subsection of our dataset which has only half the amount of rows and only a third of the columns
print("Dataset", dataset.shape)
print("Subset", ver_splits[0].shape)
````

#### Iterating

Lặp là kỹ thuật khá quen thuộc đối với lập trình.

Bây giờ chúng ta sẽ dùng vòng lặp để "ghé nhà" từng phần tử của list và "hỏi thăm" giá trị của chúng ra sao!

Ở cách này chúng ta khá "chày cối" trong việc truy xuất phần tử của mảng nhiều chiều bằng `nditer`. 

Nói như thế là bởi vì dataset mà chúng ta có là một mảng 2 chiều nhưng chúng ta đang truy xuất thuần túy của mảng một chiều nên các giá trị được in ra như một danh sách khá là khó biết được vị trí của phần tử chính xác là nơi nào.

````python
# iterating over whole dataset (each value in each row)
current_ind = 0
for x in np.nditer(dataset):
    print(x, current_ind)
    current_ind += 1
````

Để đảm bảo rằng phần tử có index dòng - cột rõ ràng, chúng ta dùng ndenumerate khi lặp trên mảng nhiều chiều:

````python
# iterating over whole dataset with indices matching the position in the dataset
for index, value in np.ndenumerate(dataset):
    print(index, value)
````

> **Chú ý:**
Dấu `^` được sử dụng để áp dụng một quy tắc regex chỉ cho phần đầu tiên của chuỗi. Nó chỉ khớp với chuỗi nếu chuỗi bắt đầu bằng mẫu regex cụ thể. Điều này cho phép bạn lọc ra các cột có tên bắt đầu bằng "2" từ DataFrame.

````python
# Lọc các quốc gia có tên bắt đầu bằng `A`
dataset.filter(regex="^A", axis=0).head()
````

````python
# Lọc các quốc gia có chứa từ "land"
dataset.filter(like="land", axis=0).head()
````

#### Reshaping

Ý tưởng phác thảo như sau:
```
Country Code    ABW    AFG    AGO   ...

        2015    577     49     20   ...
```

````python
# reshaping năm 2015 dưới dạng hàng và mã quốc gia dưới dạng cột
dataset_2015 = dataset[["Country Code", "2015"]]

dataset_2015.pivot(index=["2015"] * len(dataset_2015), columns="Country Code", values="2015")
````
