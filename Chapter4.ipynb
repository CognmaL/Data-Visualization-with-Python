## Intelligence Quotient
In this activity, we will compare IQ scores among different test groups using Seaborn.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mydata = pd.read_csv("./data/scores.csv")
```

```python
group_a = mydata[mydata.columns[0]].tolist()
group_b = mydata[mydata.columns[1]].tolist()
group_c = mydata[mydata.columns[2]].tolist()
group_d = mydata[mydata.columns[3]].tolist()
```

```python
print(group_a)
```

```python
print(group_b)
```

```python
print(group_c)
```

```python
print(group_d)
```

```python
df = pd.DataFrame({'Groups': ['Group A'] * len(group_a) + ['Group B'] * len(group_b) + ['Group C'] * len(group_c) + ['Group D'] * len(group_d),
                     'IQ score': group_a + group_b + group_c + group_d})
```

Create a box plot for each of the IQ scores of different test groups using Seaborn's boxplot function. Use the whitegrid style, set the context to talk, and remove all axes splines, except the one on the bottom. Add a title.

```python
plt.figure(dpi=150)
# Set style
sns.set_style('whitegrid')
# Create boxplot
sns.boxplot(x='Groups', y='IQ score', data=df)
# Despine
sns.despine(left=True, right=True, top=True)

# Add title
plt.title('IQ scores for different test groups')
# Show plot
plt.show()
```

## Movie comparison revisited
In this activity, we will use a bar plot to compare movie scores. You are given five movies with scores from Rotten Tomatoes. The Tomatometer is the percentage of approved Tomatometer critics who have given a positive review for the movie. The Audience Score is the percentage of users who have given a score of 3.5 or higher out of 5. Compare these two scores among the five movies.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Use pandas to read the data located in the subfolder data and transform the data into a useable format for Seaborn's barplot function.

```python
mydata = pd.read_csv("./data/movie_scores.csv")
```

```python
mydata
```

```python
movie_scores = pd.DataFrame({"Movie Title": list(mydata["MovieTitle"]) * 2,
                "Score": list(mydata["AudienceScore"]) + list(mydata["Tomatometer"]),
                "Type": ["Audience Score"] * len(mydata["AudienceScore"]) + ["Tomatometer"] * len(mydata["Tomatometer"])})
```

```python
movie_scores
```

Use Seaborn to create a visually-appealing bar plot comparing the two scores for all five movies.

```python
sns.set()
plt.figure(figsize=(10, 5), dpi=300)
# Create bar plot
ax = sns.barplot(x="Movie Title", y="Score", hue="Type", data=movie_scores)
plt.xticks(rotation=10)
# Add title
plt.title("Movies Scores comparison")
plt.xlabel("Movies")
plt.ylabel("Scores")
# Show plot
plt.show()
```

## Top 30 YouTube Channels
In this activity FacetGrid is used to show the number of subscribers and the number of views for the top 30 YouTube channels.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mydata = pd.read_csv("./data/youtube.csv")
```

```python
mydata.head()
```

```python
channels = mydata[mydata.columns[0]].tolist()
```

```python
print(channels)
```

```python
subs = mydata[mydata.columns[1]].tolist()
```

```python
views = mydata[mydata.columns[2]].tolist()
```

```python
print(subs)
```

```python
print(views)
```

```python
data = pd.DataFrame({'YouTube Channels': channels + channels, 'Subscribers in millions': subs + views, 
                     'Type': ['Subscribers'] * len(subs) + ['Views'] * len(views)})
```

Visualize the given data using a FacetGrid with two columns. The first column should show the number of subscribers for each YouTube channel, whereas the second column should show the number of views.

```python
sns.set()
g = sns.FacetGrid(data, col='Type', hue='Type', sharex=False, height=8)
g.map(sns.barplot, 'Subscribers in millions', 'YouTube Channels')
plt.show()
```

## Linear regression
In this activity, we will use a regression plot to visualize the linear relationships.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mydata = pd.read_csv("./data/anage_data.csv")
```

```python
mydata.head()
```

The given dataset is not complete. Filter the data so you end up with samples containing a body mass and a maximum longevity. Only consider samples for the class Mammalia and a body mass below 200,000.

```python
longevity = 'Maximum longevity (yrs)'
mass = 'Body mass (g)'
data = mydata[mydata['Class'] == 'Mammalia']
data = data[np.isfinite(data[longevity]) & np.isfinite(data[mass]) & (data[mass] < 200000)]
```

Create a regression plot to visualize the linear relationship of the variables.

```python
# Create figure
sns.set()
plt.figure(figsize=(10, 6), dpi=300)
# Create scatter plot
sns.regplot(x=mass, y=longevity, data=data)
# Show plot
plt.show()
```

## Water usage revisited
In this activity, we will use a tree map to visualize the water usage.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import squarify
```

Use pandas to read the data located in the subfolder data.

```python
mydata = pd.read_csv("./data/water_usage.csv")
```

```python
mydata
```

Use a tree map to visualize the water usage. Show the percentages for each tile and add a title.

```python
# Create figure
plt.figure(dpi=200)
# Create tree map
labels = mydata['Usage'] + ' (' + mydata['Percentage'].astype('str') + '%)'
squarify.plot(sizes=mydata['Percentage'], label=labels, color=sns.light_palette('green', mydata.shape[0]))
plt.axis('off')
# Add title
plt.title('Water usage')
# Show plot
plt.show()
```
